В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. 
Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?

```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame()
df['values'] = lst
df
```

```python
def transform_in_one_hot(dframe, name_column):
    sp_columns = list()

    for item in dframe[name_column]:
        if item in sp_columns:
            continue
        else:
            sp_columns.append(item)

    for name in sp_columns:
        dframe.loc[dframe[name_column] == name, name] = "1"
        dframe.loc[dframe[name_column] != name, name] = "0"
    
    dframe.drop(name_column, axis=1, inplace=True)
```

```python
transform_in_one_hot(df, 'values')
df
```
